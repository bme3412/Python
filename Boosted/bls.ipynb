```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, urlretrieve, quote
from urllib.parse import urljoin

url = "https://www.bls.gov/cpi/tables/supplemental-files/home.htm"

soup = BeautifulSoup(requests.get(url).content, "html.parser")


# need to put into a separate folder

for link in soup.select('a[href*=".xlsx"]'):
    href = str('https://www.bls.gov')+link.get('href')

    # Make sure it has one of the correct extensions
    if not any(href.endswith(x) for x in ['.xlsx']):
        continue
    
    filename = href.rsplit('/', 1)[-1]
    print("Downloading %s to %s..." % (href, filename) )
    urlretrieve(href, filename)
    print("Done.")

print('Extracted all files')
```
