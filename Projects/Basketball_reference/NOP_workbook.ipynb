```python
season = 2019
appended_data_NOP = []

# NOP
while season <= 2021:
    df_NOP = get_roster('NOP', season)
    df_NOP['Season'] = season
    df_NOP['Team'] = 'NOP'
    appended_data_NOP.append(df_NOP)
    season = season + 1
appended_data_NOP = pd.concat(appended_data_NOP)
```

```python
NOP_df_all = appended_data_NOP.copy()
NOP_df_all_csv = NOP_df_all.to_csv('/Users/brendan/Desktop/Python/NBA_API/NOP/NOP_players_all.csv', index=False)
NOP_df_all_csv_io = pd.read_csv('/Users/brendan/Desktop/Python/NBA_API/NOP/NOP_players_all.csv')
NOP_df_all_csv_io.head()
```
