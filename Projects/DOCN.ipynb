```python
ax = df[['Date', 'Revenue', 'Gross Profit - nonGAAP']].plot(x='Date',kind='barh', stacked=True, figsize=(15,10))
# format xtick-labels with list comprehension
ax.set_yticklabels([x.strftime("%Y-%m-%d") for x in df.Date], rotation=45)

plt.gca().invert_yaxis()
plt.show()
```
