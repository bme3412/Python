```python
import pandas as pd
import numpy as np
```

```python
df_t0 = pd.DataFrame(np.random.randint(5,1000, size=(100,5)),columns=(['Price','Market Cap', 'Price Target','Date', '% YTD']))

min_date = pd.to_datetime('2021-10-30')
max_date = pd.to_datetime('2022-01-1')

d = (max_date - min_date).days + 1
```

```python
df_t0['Price'] = np.random.randint(25,1000, size=len(df_t0))
df_t0['Market Cap'] = np.random.randint(500,50000, size=len(df_t0))
df_t0['% YTD'] = np.random.randint(-10,70, size=len(df_t0))
df_t0['Date'] = min_date + pd.to_timedelta(pd.np.random.randint(d,size=len(df_t0)), unit='d')
```

```python
df_t0 = df_t0.reset_index().head()
```

```python
df_t1 = pd.DataFrame(np.random.randint(5,1000, size=(100,5)),columns=(['Price','Market Cap', 'Price Target','Date', '% YTD']))

min_date = pd.to_datetime('2021-01-02')
max_date = pd.to_datetime('2022-01-13')

d = (max_date - min_date).days + 1

df_t1['Price'] = np.random.randint(25,1000, size=len(df_t1))
df_t1['Market Cap'] = np.random.randint(500,50000, size=len(df_t1))
df_t1['% YTD'] = np.random.randint(-10,70, size=len(df_t1))
df_t1['Date'] = min_date + pd.to_timedelta(pd.np.random.randint(d,size=len(df_t1)), unit='d')
```

```python
df_t1 = df_t1.reset_index().head()
```

```python
comparison_vals = df_t0.values == df_t1.values
row,cols = np.where(comparison_vals ==False )
for item in zip(row,cols):
    df_t1.iloc[item[0], item[1]] = '{} {}'.format(df_t0.iloc[item[0],item[1]], df_t1.iloc[item[0],item[1]])
```

```python
def highlight_diff(data, color='coral'):
    attr = 'background-color: {}'.format(color)
    other = data.xs('First', axis='columns', level=-1)
    return pd.DataFrame(np.where(data.ne(other, level=0), attr, ''),
                        index=data.index, columns=data.columns)

df_final.style.apply(highlight_diff, axis=None)
```
