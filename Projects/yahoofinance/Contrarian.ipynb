```python
df = price_df.set_index('symbol')
def is_contrarian(symbol):
    price = df.loc[f'{symbol}', 'currentPrice']
    price_high = df.loc[f'{symbol}', 'fiftyTwoWeekHigh']
    price_low = df.loc[f'{symbol}', 'fiftyTwoWeekLow']
    price_band = (price_high - price_low)
    if price < (price_low + (.3*(price_band))):
        return f'{symbol}'
    else:
        return 'OK'
```

```python
is_contrarian('LRCX')
```

```python
contrarian_tickers = map(is_contrarian, symbol_list)
ticks = [i for i in symbol_list if is_contrarian(i) != 'OK' ]


```

## Price targets

```python
px_target_cols = ['symbol', 'sector', 'industry','currentPrice', 'targetLowPrice', 'targetMedianPrice','numberOfAnalystOpinions',
            'targetMeanPrice', 'targetHighPrice', 'recommendationMean']
```

```python
appended_data = []
for ticker in ticks:
    data = yf.Ticker(ticker)
    df = pd.DataFrame([data.info])
    px_target = df[px_target_cols]
    appended_data.append(px_target)

contrarian_df = pd.DataFrame(np.row_stack(appended_data))
contrarian_df.columns =  ['Ticker', 'Sector', 'Industry','Price', 'Low Price', 'Median Price','# of Ratings',
            'Mean Price', 'High Price', 'Avg Rating']
contrarian_df = contrarian_df[['Ticker', 'Sector', 'Industry','Price', 'Low Price', 'Median Price', 'Mean Price', 'High Price',
                 '# of Ratings', 'Avg Rating']]
contrarian_df
```

```python
data_merge.columns = ['Ticker', 'Market Cap', 'FiftyTwoWeekHigh', 'FiftyTwoWeekLow']
data_merge
```

```python
merged_cont_df = contrarian_df.merge(data_merge, how = 'inner', on = ['Ticker'])
merged_cont_df = merged_cont_df[['Ticker', 'Market Cap', 'Sector', 'Industry', 'Price', 'FiftyTwoWeekHigh', 'FiftyTwoWeekLow',
                  'Low Price', 'Median Price', 'Mean Price', 'High Price', '# of Ratings', 'Avg Rating']]

```

```python
merged_cont_df['Contrarian %'] = (merged_cont_df['Price'] - merged_cont_df['FiftyTwoWeekLow'])/(merged_cont_df['FiftyTwoWeekHigh'] - merged_cont_df['FiftyTwoWeekLow'])*100
                                                  
                                                                                               
```
