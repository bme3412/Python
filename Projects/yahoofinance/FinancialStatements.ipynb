```python
from ticker_convert import ticker_convert

#SP_500_tech_df = []

files =glob.glob('*csv')

SP500_tech_tickers_list = []
for i in files:
    x = i.replace(".csv", "")
    SP500_tech_tickers_list.append(x)
SP500_tech_tickers_list

#tickers_df = map(ticker_convert,SP500_tech_tickers_list)

def tickers_df(ticker):
    for tick in ticker:
        df_raw = yf.Ticker(f'{ticker}').info
        df = pd.DataFrame([df_raw])
    return df
        
```

```python
tickers_df('MSFT')
```

```python
len(SP500_tech_tickers_list)
SP500_tech_tickers_list_1 = SP500_tech_tickers_list[:10]
SP500_tech_tickers_list_2 = SP500_tech_tickers_list[10:20]
SP500_tech_tickers_list_3 = SP500_tech_tickers_list[20:30]
SP500_tech_tickers_list_4 = SP500_tech_tickers_list[30:40]
SP500_tech_tickers_list_5 = SP500_tech_tickers_list[40:50]
SP500_tech_tickers_list_6 = SP500_tech_tickers_list[50:60]
SP500_tech_tickers_list_7 = SP500_tech_tickers_list[60:70]
SP500_tech_tickers_list_8 = SP500_tech_tickers_list[70:80]
SP500_tech_tickers_list_9 = SP500_tech_tickers_list[80:90]
SP500_tech_tickers_list_10 = SP500_tech_tickers_list[90:101]
```
