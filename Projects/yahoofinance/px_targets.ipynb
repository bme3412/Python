```python
amzn = yf.Ticker('AMZN')
amzn_info = pd.DataFrame([amzn.info])
amzn_info
```

```python
ticker_list = [['AAPL', 'MSFT', 'FB', 'AMZN', 'GOOGL']]

def stock_daily(ticker_list):
    start = pd.Timestamp('now').normalize() - pd.offsets.Day(1000)
    end = pd.Timestamp('now').normalize() + pd.offsets.BusinessDay(0)
    return yf.download(ticker_list, start=start, end=end).Close

def ticker_data(list):
    data = []
    for ticks in list:
        data.append(stock_daily(ticks))
    return data

res = ticker_data(ticker_list)
dataframes = [pd.DataFrame(ticker_list) for lst in res]
print(dataframes[0])

df = pd.concat(map(stock_daily, ticker_list), axis=1)

# I transposed for prettier printing
df.T
```
