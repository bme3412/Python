```python
amzn_info = pd.DataFrame([amzn.info])
amzn_info
```

```python
amzn_info.columns.to_list()
```

```python
px_target_cols = ['symbol', 'sector', 'industry','currentPrice', 'targetLowPrice', 'targetMedianPrice','numberOfAnalystOpinions',
            'targetMeanPrice', 'targetHighPrice', 'recommendationMean']

```

```python
amzn_px_target = amzn_info[px_target_cols]
amzn_px_target
```

```python
amzn_px_target['% to Median'] = amzn_px_target['targetMedianPrice']/amzn_px_target['currentPrice']-1
amzn_px_target['% Upside'] = amzn_px_target['targetHighPrice']/amzn_px_target['currentPrice']-1
amzn_px_target['% Downside'] = amzn_px_target['targetLowPrice']/amzn_px_target['currentPrice']-1
amzn_px_target
```
