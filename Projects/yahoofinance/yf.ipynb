```python
ticker_list = [['AAPL', 'MSFT', 'FB', 'AMZN', 'GOOGL']]

def stock_daily(ticker_list):
    start = pd.Timestamp('now').normalize() - pd.offsets.Day(1000)
    end = pd.Timestamp('now').normalize() + pd.offsets.BusinessDay(0)
    return yf.download(ticker_list, start=start, end=end).Close

def ticker_data(list):
    data = []
    for ticks in list:
        data.append(stock_daily(ticks))
    return data



res = ticker_data(ticker_list)
dataframes = [pd.DataFrame(ticker_list) for lst in res]
print(dataframes[0])

df = pd.concat(map(stock_daily, ticker_list), axis=1)

# I transposed for prettier printing
df.T
```

```python
df.head()
```

```python
df.columns
```

```python
df['AAPL_%1D'] = df['AAPL'].pct_change() *100
df['AMZN_%1D'] = df['AMZN'].pct_change() *100
df['FB_%1D'] = df['FB'].pct_change() *100
df['GOOGL_%1D'] = df['GOOGL'].pct_change() * 100
df['MSFT_%1D'] = df['MSFT'].pct_change() * 100

df['AAPL_Return'] = df['AAPL_%1D'].cumsum()
df['AMZN_Return'] = df['AMZN_%1D'].cumsum()
df['FB_Return'] = df['FB_%1D'].cumsum()
df['GOOGL_Return'] = df['GOOGL_%1D'].cumsum()
df['MSFT_Return']= df['MSFT_%1D'].cumsum()
```

```python
df = df.dropna().reset_index()
df.head(3)
```

```python
ticker_list
```

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(20,20))
x = df['Date']
y1 = df['AAPL_Return']
y2 = df['AMZN_Return']
y3 = df['FB_Return']
y4 = df['GOOGL_Return']

for Y in [y1, y2, y3, y4]:
    ax.plot(x, Y)
plt.title('MegaCap Tech Trailing 1000 Days')
ax.legend(['AAPL','AMZN', 'FB', 'GOOGL'],fancybox=True,loc='best', shadow=True,borderpad=2)
plt.show()
```

```python
tech = ndx_df[ndx_df['Sector']=='Information Technology']
tech.head()
```

```python
semiconductors = tech[tech['Industry']=='Semiconductors']
semiconductors.head()
```

```python
semi_tickers = semiconductors['Ticker']
semi_ticker_list = semi_tickers.to_list()
semi_ticker_list
```

```python
type(semi_ticker_list)
ticker_list = ' '.join(semi_ticker_list)
ticker_list
```

```python
#semi_ticker_list = [['AAPL', 'MSFT', 'FB', 'AMZN', 'GOOGL']]
ticker_list_semi = semi_ticker_list

def stock_daily(ticker_list_semi):
    start = pd.Timestamp('now').normalize() - pd.offsets.Day(200)
    end = pd.Timestamp('now').normalize() + pd.offsets.BusinessDay(0)
    return yf.download(ticker_list, start=start, end=end).Close

def ticker_data(list):
    data = []
    for ticks in list:
        data.append(stock_daily(ticks))
    return data



res = ticker_data(ticker_list_semi)
dataframes = [pd.DataFrame(ticker_list_semi) for lst in res]
print(dataframes[0])

df_semi = pd.concat(map(stock_daily, ticker_list_semi), axis=1)

# I transposed for prettier printing
df_semi.T[:11].reset_index()
```

```python
df_semi = df_semi.T.drop_duplicates().T
df_semi = df_semi.reset_index()
```

```python
df_semi['ADI_%1D'] = df_semi['ADI'].pct_change() *100
df_semi['AMD_%1D'] = df_semi['AMD'].pct_change() *100
df_semi['AVGO_%1D'] = df_semi['AVGO'].pct_change() *100
df_semi['INTC_%1D'] = df_semi['INTC'].pct_change() * 100
df_semi['MCHP_%1D'] = df_semi['MCHP'].pct_change() * 100
df_semi['MU_%1D'] = df_semi['MU'].pct_change() *100
df_semi['NVDA_%1D'] = df_semi['NVDA'].pct_change() *100
df_semi['NXPI_%1D'] = df_semi['NXPI'].pct_change() *100
df_semi['QCOM_%1D'] = df_semi['QCOM'].pct_change() * 100
df_semi['SWKS_%1D'] = df_semi['SWKS'].pct_change() * 100
df_semi['TXN_%1D'] = df_semi['TXN'].pct_change() *100


df_semi['ADI_Return'] = df_semi['ADI_%1D'].cumsum()
df_semi['AMD_Return'] = df_semi['AMD_%1D'].cumsum()
df_semi['AVGO_Return'] = df_semi['AVGO_%1D'].cumsum()
df_semi['INTC_Return'] = df_semi['INTC_%1D'].cumsum()
df_semi['MCHP_Return']= df_semi['MCHP_%1D'].cumsum()
df_semi['MU_Return'] = df_semi['MU_%1D'].cumsum()
df_semi['NVDA_Return'] = df_semi['NVDA_%1D'].cumsum()
df_semi['NXPI_Return'] = df_semi['NXPI_%1D'].cumsum()
df_semi['QCOM_Return'] = df_semi['QCOM_%1D'].cumsum()
df_semi['SWKS_Return']= df_semi['SWKS_%1D'].cumsum()
df_semi['TXN_Return']= df_semi['TXN_%1D'].cumsum()
```

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(20,20))
x = df_semi['Date']
y1 = df_semi['ADI_Return']
y2 = df_semi['AMD_Return']
y3 = df_semi['AVGO_Return']
y4 = df_semi['INTC_Return']
y5 = df_semi['MCHP_Return']
y6 = df_semi['MU_Return']
y7 = df_semi['NVDA_Return']
y8 = df_semi['NXPI_Return']
y9 = df_semi['QCOM_Return']
y10 = df_semi['SWKS_Return']
y11 = df_semi['TXN_Return']



for Y in [y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11]:
    ax.plot(x, Y)
plt.title('Semis Trailing 10 Days')
ax.legend(['ADI','AMD', 'AVGO', 'INTC','MCHP', 'MU', 'NVDA', 'NXPI', 'QCOM', 'SWKS', 'TXN'],fancybox=True,loc='best', shadow=True,borderpad=2)
plt.show()
```
