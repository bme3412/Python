```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
x = list(range(11))
y0, y1, y2 = x, [10-i for i in x], [abs(i-5) for i in x]
```

## Rows and Columns

```python
from bokeh.layouts import row

# create a new plot
s1 = figure(width=300, plot_height=300)
s1.circle(x,y0, size=10, color='navy', alpha=0.5)

# create another one
s2 = figure(width=300, height=300)
s2.triangle(x,y1, size=10, color='firebrick', alpha=0.5)

# create another
s3 = figure(width=300, height=300)
s3.square(x,y2,size=10, color='green', alpha=0.5)

# show the results in a row
show(row(s1, s2, s3))
```

## Grid plots

```python
from bokeh.layouts import gridplot

# create a new plot
s1 = figure(width=250, plot_height=250)
s1.circle(x, y0, size=10, color='navy', alpha=0.5)

# create another one
s2 = figure(width=250, height=250)
s2.triangle(x, y1, size=10, color='firebrick', alpha=0.5)

# create another
s3 = figure(width=250, height=250)
s3.square(x, y2, size=10, color='green', alpha=0.5)

# put all the plots in a gridplot
p = gridplot([[s1, s2], [s3, None]], toolbar_location=None)

show(p)
```
