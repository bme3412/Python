```python
product_title = []
prices = []

def df_scrape_page1(i):
    url = i 
    buy_it_now = '?rt=nc&LH_BIN=1'
    url_sold = url+buy_it_now 
    print(url)
    page = requests.get(url)
    soup = bs(page.content, "lxml")
        #date = url.split('/')[28:36]
        #print(date)

    items = soup.find_all('div', class_='s-item__info clearfix')
    #print(items)
    for item in items:
        title_all = soup.find_all('h3', class_='s-item__title')
        for title in title_all:
            product_title.append(title.text)
        price_all = soup.find_all('span', class_="s-item__price")
        for price in price_all:
            prices.append(price.text)
    #y = df_scrape(df_2019_test['Link']) 
    x = product_title, prices

    result = pd.DataFrame(tuple(zip(x[0],x[1])),columns=['Product','Price'])
    date_ = url
    #print(date_)
    date = date_[28:36]
        #print(date_)
    result['Date'] = date
    new_df = result.copy()
    new_df = new_df.drop_duplicates(subset='Product')
    #return new_df
    new_df.to_csv(f'iPhone_{date}_page1.csv')



    
```
