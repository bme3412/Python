## 1 day

```python
SP500_metrics = data2[data2['ticker'].isin(tickers_SP500)]
SP500_metrics_06_01_2022 = SP500_metrics[SP500_metrics['date']=='2022-06-01']

SP500_prices = data3[data3['ticker'].isin(tickers_SP500)]
SP500_prices_06_01_2022 = SP500_prices[SP500_prices['date']=='2022-06-01']

SP500_06_01_2022 = pd.merge(SP500_metrics_06_01_2022, SP500_prices_06_01_2022,on='ticker')
SP500_06_01_2022 = SP500_06_01_2022.drop(['lastupdated_x','lastupdated_y', 'date_y'],axis=1)
SP500_06_01_2022.head()
```

## 1 day return

```python
SP500_prices_06_01_2022 = SP500_06_01_2022[['ticker', 'date_x', 'closeadj']]
SP500_prices_06_02_2022 = SP500_06_02_2022[['ticker', 'date_x', 'closeadj']]
SP500_1day_06_02_2022 = pd.concat([SP500_prices_06_01_2022, SP500_prices_06_02_2022])
SP500_1day_06_02_2022['1_day_%'] = SP500_1day_06_02_2022.groupby(['ticker'])['closeadj'].pct_change()*100
SP500_1day_06_02_2022= SP500_1day_06_02_2022.dropna()
```

```python
SP500_1day_06_02_2022_merged = SP500_1day_06_02_2022.merge(df_SP500,how='inner',on=['ticker'])
SP500_1day_06_02_2022_merged.head()
```

## Market cap wgt return

```python
SP500_06_01_2022_mcap = SP500_06_01_2022.merge(df_SP500,how='inner',on=['ticker'])
SP500_06_01_2022_mcap= SP500_06_01_2022_mcap[['date_x', 'ticker','Company','Sector','marketcap','ev','evebit','pb','pe','ps',
                       'open','high','low','close','volume','closeadj','closeunadj']]
SP500_06_01_2022_mcap.to_csv('SP500_06_01_2022.csv')

```

```python
SP500_06_02_2022_mcap = SP500_06_02_2022.merge(df_SP500,how='inner',on=['ticker'])
SP500_06_02_2022_mcap= SP500_06_02_2022_mcap[['date_x', 'ticker','Company','Sector','marketcap','ev','evebit','pb','pe','ps',
                       'open','high','low','close','volume','closeadj','closeunadj']]
SP500_06_02_2022_mcap.to_csv('SP500_06_02_2022.csv')
```

```python
SP500_06_02_mcap_chg = pd.concat([SP500_06_01_2022_mcap, SP500_06_02_2022_mcap])
SP500_06_02_mcap_chg['1_day_%'] = SP500_06_02_mcap_chg.groupby(['ticker'])['marketcap'].pct_change()*100
SP500_06_02_mcap_chg.dropna().head()
```

```python
SP500_06_01_2022_sector = SP500_06_01_2022_mcap.groupby(["Sector"])['marketcap'].agg('sum').to_frame()
SP500_06_02_2022_sector = SP500_06_02_2022_mcap.groupby(["Sector"])['marketcap'].agg('sum').to_frame()
SP500_06_02_2022_sector
```

```python
SP500_06_02_2022_sector_chg = pd.concat([SP500_06_01_2022_sector, SP500_06_02_2022_sector])
SP500_06_02_2022_sector_chg['1_day_%'] = SP500_06_02_2022_sector_chg.groupby(['Sector'])['marketcap'].pct_change()*100

SP500_06_02_2022_sector_chg.dropna()

```

```python
SP500_06_02_mcap_chg_sector = pd.merge(SP500_06_02_mcap_chg,SP500_06_02_2022_sector_chg,how='inner',on=['Sector']).dropna()
len(SP500_06_02_mcap_chg_sector)                         
                                        
```

```python
SP500_06_02_mcap_chg_sector['Active'] = SP500_06_02_mcap_chg_sector['1_day_%_x'] - SP500_06_02_mcap_chg_sector['1_day_%_y']
SP500_06_02_mcap_chg_sector.head()
```
