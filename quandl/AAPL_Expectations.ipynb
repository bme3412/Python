```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
raw_df_2019 = pd.read_csv('AAPL_2019.csv')
raw_df_2020 = pd.read_csv('AAPL_2020.csv')
```

```python
raw_df_2020.dtypes
```

```python
raw_df_2020['EPS'] = raw_df_2020['close']/raw_df_2020['pe']
raw_df_2020['date'] = pd.to_datetime(raw_df_2020['date'])
raw_df_2019['date'] = pd.to_datetime(raw_df_2019['date'])
```

```python
raw_df_2020['date_1yr_ago'] = raw_df_2020['date'] -timedelta(days=365)
```

```python
len(raw_df_2020)
```

```python
len(raw_df_2019)
```

```python
AAPL_19_20_df = raw_df_2020.merge(raw_df_2019,how='left',on=['ticker'])
len(AAPL_19_20_df)
```

```python
AAPL_19_20_df = AAPL_19_20_df.drop_duplicates(subset=['date_x','date_y'], keep=False)
AAPL_19_20_df
```

```python
AAPL_19_20_df['date_x_day'] = AAPL_19_20_df['date_x'].dt.day_name()
AAPL_19_20_df['date_1_yr'] = AAPL_19_20_df['date_1yr_ago'].dt.day_name()
AAPL_19_20_df.tail()
```

```python
AAPL = pd.DataFrame()

AAPL['2019'] = pd.bdate_range("2019-01-01", "2019-12-31")
AAPL['2020'] = pd.bdate_range("2020-01-02", "2020-12-31")
AAPL['2021'] = pd.bdate_range("2021-01-01", "2021-12-31")


AAPL.columns
```

```python
x_2019 = np.linspace(1,len(AAPL['2019']),num=len(AAPL['2019'])).reshape([len(x_2019),1])
x_2020 = np.linspace(1,len(AAPL['2019']),num=len(AAPL['2020'])).reshape([len(x_2020),1])
x_2021 = np.linspace(1,len(AAPL['2019']),num=len(AAPL['2021'])).reshape([len(x_2021),1])
```

```python
AAPL['BD_2019'] = x_2019
AAPL['BD_2020'] = x_2020
AAPL['BD_2021'] = x_2021
AAPL['Ticker'] = 'AAPL'


#AAPL['2020_px'] = 
#AAPL['2020_px'] = 


AAPL.head()
```
