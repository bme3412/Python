## AMD

```python
amd_raw = pd.read_excel('AMD_model.xlsx', sheet_name='Python')
amd_raw.columns = amd_raw.iloc[0]
amd_raw.iloc[1:]
amd_raw.dropna(axis = 0, how = 'all', inplace = True)
amd_raw.dropna(axis = 1, how = 'all', inplace = True)
```

```python
amd_raw = amd_raw.rename(columns={np.nan: 'Metric'})
#amd_raw['Ticker'] = 'AMD'
amd_raw_df = amd_raw.copy()
```

```python
subset_list = ['Ticker','Revenue','Revenue % y/y','Gross Profit','Gross Profit % y/y',
              'Gross Profit % of revenue','Operating Profit','Operating Profit % y/y',
              'FCF','FCF % y/y','Inventory']
```

```python
amd_raw_df = amd_raw.set_index('Metric')
amd_raw_df.head()
```

```python
amd_df = amd_raw_df.loc[[np.nan,'Revenue','Revenue % y/y','Gross Profit','Gross Profit % y/y',
              'Gross Profit % of revenue','Operating Profit','Operating Profit % y/y',
              'FCF','FCF % y/y'],:]
```

```python
amd_transposed_df = amd_df.T.copy()
amd_transposed_df.columns = ['Quarter','Revenue','Revenue % y/y','Gross Profit','Gross Profit % y/y',
                            'Gross Profit Margin %','Operating Profit','Operating Profit % y/y','FCF',
                            'FCF % y/y']
amd_transposed_df['Ticker'] = 'AMD'
amd_transposed_df.to_excel('AMD_transposed_df.xlsx')
```
