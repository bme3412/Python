```python
import pandas as pd
import glob
from datetime import datetime
```

```python
files =glob.glob('*csv')
```

```python
from ticker_convert import ticker_convert
```

```python
files
SP500_tickers_list = []
for i in files:
    x = i.replace(".csv", "")
    SP500_tickers_list.append(x)
SP500_tickers_list
```

```python
tickers_df = map(ticker_convert,SP500_tickers_list)
SP500_master_df = []
for i in tickers_df:
    df = pd.DataFrame(i)
    SP500_master_df.append(df)
SP500_master_df = pd.concat(SP500_master_df)
```

```python
today_date = datetime.now()
date = today_date.strftime('%m/%d/%Y %H:%M:%S')
SP500_master_df['Date'] = date
SP500_master_df.to_csv('output/SP500_master_df.csv')
```

```python
SP500_master_df.columns
```

```python
raw_df = pd.read_csv('output/SP500_master_df.csv')
raw_df.columns
raw_df = raw_df.drop('Unnamed: 0', 1)
raw_df = raw_df[['Date','symbol', 'marketCap', 'enterpriseValue', 'currentPrice',
       'fiftyTwoWeekHigh', 'fiftyTwoWeekLow', 'earningsGrowth',
       'enterpriseToRevenue', 'enterpriseToEbitda', 'forwardEps',
       'trailingEps', 'priceToSalesTrailing12Months', 'forwardPE',
       'dividendYield']]
raw_df
```

```python
date_file =today_date.strftime('%m_%d_%Y')
raw_df.to_csv(f'output/SP500_master_df_{date_file}.csv')
```
