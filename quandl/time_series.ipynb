```python
import pandas as pd
from datetime import datetime, timedelta
from pandas.tseries.offsets import *
```

```python
df_2018 = pd.read_csv('SP_500_2018_sort.csv')
df_2019 = pd.read_csv('SP_500_2019_sort.csv')
```

```python
NVDA_df = df_2018[df_2018['Ticker']=='NVDA']
NVDA_df.drop('Unnamed: 0',axis=1,inplace=True)

NVDA_df['date'] = pd.to_datetime(NVDA_df['date'])
NVDA_df['date_T365'] = NVDA_df['date']+BDay(260)
```
